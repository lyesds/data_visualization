### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
tips.info()
tips
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips.total_bill, 20)
plt.xlabel('Total bill amount ($)')
plt.ylabel('Frequency')
plt.title('Total bill ($) histogram')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.distplot(tips.total_bill)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips.total_bill, tips.tip)
plt.ylabel('Tip amount ($)')
plt.xlabel('Bill amount ($)')
plt.title('Tips and bills amounts ($)')
plt.show()
```

```python
import scipy.stats as stats
sns.jointplot(data=tips, x='total_bill', y='tip')
r, p = stats.pearsonr(tips.total_bill, tips.tip)
r, p
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips, vars=["total_bill", "tip", "size"])
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.pairplot(tips, vars=["total_bill", "day"])
```

```python
sns.catplot(data=tips, x=tips.day, y=tips.total_bill, kind='swarm')
```

```python
sns.catplot(data=tips, x=tips.day, y=tips.total_bill)
plt.grid(True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
# plt.scatter(x=tips.tip, y=tips.day, c=tips.sex)
sns.stripplot(x="tip", y="day", data=tips, hue='sex')
plt.title('Tips amounts ($) by day, colored by sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.catplot(x="day", y="total_bill", hue="time",
            kind="box", dodge=True, data=tips)
```

```python
sns.boxplot(x="day", y="total_bill",
            hue="time", palette=["b", "g"],
            data=tips)
plt.ylim([0,60])
plt.grid(True)
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
sns.displot(data=tips, x="tip", col="time")
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="sex", hue="smoker", fit_reg=False, palette=['b', 'g'])
```
